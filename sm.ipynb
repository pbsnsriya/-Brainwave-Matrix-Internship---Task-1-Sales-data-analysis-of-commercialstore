```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(dataa['Total'], bins=30, kde=True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot total sales by product line
plt.figure(figsize=(12, 8))
sns.boxplot(x='Product line', y='Total', data=dataa)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Date' to datetime format
dataa['Date'] = pd.to_datetime(dataa['Date'], format='%m/%d/%Y')

# Plot total sales over time
plt.figure(figsize=(14, 7))
dataa.groupby('Date')['Total'].sum().plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Plot sales by payment method
plt.figure(figsize=(10, 6))
sns.boxplot(x='Payment', y='Total', data=dataa)
plt.title('Sales by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Sales')
plt.show()
```
